```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
idade_mae = sinasc['IDADEMAE']

# Crie o boxplot usando a função boxplot do matplotlib
plt.boxplot(idade_mae)


# Mostre o boxplot
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
peso_bebe = sinasc['PESO']
```

```python
plt.boxplot(peso_bebe)


# Mostre o boxplot
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
idade_mae = sinasc['IDADEMAE']
```

```python
plt.hist(idade_mae, bins=5)

plt.xlabel('Valores')
plt.ylabel('Frequência')

# Defina um título para o gráfico
plt.title('Histograma da Idade da Mãe')

# Mostre o gráfico
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.hist(peso_bebe, bins=10)

# Defina rótulos para o eixo x e y
plt.xlabel('Valores')
plt.ylabel('Frequência')

# Defina um título para o gráfico
plt.title('Histograma do peso do bebê')

# Mostre o gráfico
plt.show()
```
